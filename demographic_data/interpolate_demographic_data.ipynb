```python
df_demo = pd.read_csv('demo.csv')
```

```python
df_dd5 = pd.read_csv('dd5v1.csv')
```

```python
#df['zip_code'] = df['zip_code'].astype(str)
df_demo['zip_code'] = df_demo['zip_code'].astype(str)
df_dd5['zip_code'] = df_dd5['variable'].astype(str)

#df['year'] = df['year'].astype(str)
df_demo['year'] = df_demo['year'].astype(str)
df_dd5['year'] = df_dd5['year'].astype(str)

df_dd5 = df_dd5.drop(columns='variable')
```

```python
df_demo.shape
```

```python
yrs = [str(i) for i in range(1991,2011)]
#yrs
```

```python
df_dd5.shape
```

```python
# filter for pre-2011
df_dd5 = df_dd5[df_dd5['year'].isin(yrs) == True]
df_dd5.shape
```

```python
df_dd5.head()
```

```python
# d = df_dd5[df_dd5.zip_code == '89010']
# f = df_demo[df_demo.zip_code == '89010']
```

```python
df = pd.concat([df_dd5,df_demo])
```

```python
df.replace(0, np.nan, inplace=True)
df.replace(0.0, np.nan, inplace=True)
```

```python
df.shape
```

```python
df.year.value_counts()
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
zip_codes = set(list(df.zip_code))
len(zip_codes)
```

```python
# d = df[df.zip_code == '89010']
# d.sort_values('year')
# d
```

```python
df_final = pd.DataFrame()
for z in zip_codes:
    print(z)
    d = df[df.zip_code == z]
    d = d.sort_values('year')
    d = d.interpolate()
    df_final = pd.concat([df_final, d])
```

```python
df_final.fillna(method='backfill',inplace=True)
df_final = df_final.round(1)
```

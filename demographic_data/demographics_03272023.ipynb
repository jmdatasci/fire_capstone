```python
zipcodes=dm.loc[:,"zip_code"].unique()
cols=['households_count', 'mean_income_per_household',
       'median_income_per_household', 'percent_high_school_grad',
       'percent_bach_deg_grad', 'total_population', 'percent_pop_age_under_5',
       'percent_pop_age_5to9', 'percent_pop_age_10to14',
       'percent_pop_age_15to19', 'percent_pop_male_age_under_5',
       'percent_pop_male_age_5to9', 'percent_pop_male_age_10to14',
       'percent_pop_male_age_15to19', 'percent_pop_female_age_under_5',
       'percent_pop_female_age_5to9', 'percent_pop_female_age_10to14',
       'percent_pop_female_age_15to19']
```

```python
#select year 2011
#reset the index
#drop the year column
#set the index to be the zipcode
dm1=dm.set_index("year")
dmy=dm1.loc['2011']
dmy1=dmy.reset_index()
dmy2=dmy1.drop('year',axis=1)
dmy2.set_index("zip_code",inplace=True)
```

```python
#compute statewide values which will be used to generate ratios(multipliers)
#from collections import defaultdict
#d1=defaultdict(list)
k=[]
j=[]
d1=pd.DataFrame()
for i in cols:
    if i in ['households_count','total_population']:
        i_m=dmy2.loc[:,i].sum()
        k.append(i)
        j.append(i_m)
    else:
        i_m=dmy2.loc[:,i].mean()
        k.append(i)
        j.append(i_m)
        
d3=dict(zip(k,j))
d4=pd.Series(d3)
```

```python
#compute ratio based on the state wide variables
#Ratio is computed as (zipcode value/California Value) - it will be used as a multiplier
zratio=dmy2/d4
```
